```python
from sklearn.preprocessing import LabelEncoder

sf=pd.read_csv('emp.csv')
sf

label_enco=LabelEncoder()
sf['Senior Management'].value_counts()
sf['Senior Management'].nunique()


labels = label_enco.fit_transform(sf.Gender)

sf['Target'] = labels
sf
```
